# Plotting with matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = range(0, 10)
y = [i ** 2 for i in x]

# Basic plot
plt.plot(x,y)
```

```python
# Add title and axis labels
plt.plot(x,y)
plt.title('x-square')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# More than one plot
plt.subplot(1,2,1) # nrows, ncols, index of the next plot starting with index 1 from the top left and increasing to the right
plt.plot(x, y, 'r--') # 'r' stands for red, '--' stands for dash
plt.title('x-squared')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
plt.title('square root x');  # insted of plt.show() you can also use a semicolon to show the plot
```

### OOP approach

```python
x = range(0, 10)
y = [i ** 2 for i in x]

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1]) # left, bottom, width, height (range 0 to 1)
axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Simple x-squared')
```

```python


fig = plt.figure()

axes1 = fig.add_axes([0, 0, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.1, 0.4, 0.4, 0.3]) # insert axes: left and bottom of the lower-left corner, width, height

# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('x-squared')

# insert
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('square root x ')
```

## Stock market application

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://osf.io/4pgrf/download')
```

```python
df.head()
```

```python
print(f'Our data has {df.shape[0]} observations.')
```

```python
df.describe()
```

```python
df.columns
```

```python
# Date column: correct data type?
df["ref.date"].describe()
```

```python
# No; convert to datetime 
df['ref.date'] = pd.to_datetime(df['ref.date'])
```

```python
df["ref.date"].describe()
```

```python
# Which tickers do we have?
df.ticker.unique()
```

```python
# Balanced panel?
df.ticker.value_counts()
```

```python
# 2 ways to convert a Pandas column to list
date = list(df[df.ticker == 'MSFT']['ref.date'])
price = df[df.ticker == 'MSFT']['price.close'].tolist()
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])
ax.set_title('MSFT Daily Closing Prices')
ax.plot(date, price)
```

```python
# Add some additional elements to plot
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])
ax.set_title('MSFT Daily Closing Prices')
ax.plot(date, price, label = 'MSFT')
plt.legend(loc = 'upper left')
ax.set_ylim(0,180)
```

```python
# Add line corresponding to average over sample period
date = df[df.ticker == 'MSFT']['ref.date'].tolist()
price = df[df.ticker == 'MSFT']['price.close'].tolist()
meanprice = np.mean(price)
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])
ax.set_title('MSFT Daily Closing Prices')
ax.plot(date, price, label = 'MSFT')
ax.hlines(y = meanprice, xmin = date[0], xmax = date[-1], linestyle = '--', label = 'avg')
plt.legend(loc = 'upper left')
```

```python
# In log-scale
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])
ax.set_title('MSFT Daily Closing Prices')
ax.plot(date, price, label = 'MSFT')
ax.hlines(y = meanprice, xmin = date[0], xmax = date[-1], linestyle = '--', label = 'avg')
ax.set_yscale('log')
plt.legend(loc = 'upper left')
```

```python
# Plotting two series with different y-axis
date = df[df.ticker == 'MSFT']['ref.date'].tolist()
price_msft = df[df.ticker == 'MSFT']['price.close'].tolist()
price_aapl = df[df.ticker == 'AAPL']['price.close'].tolist()


fig, ax1 = plt.subplots(figsize = (10,6))

ax1.plot(date, price_msft, color = 'k')
ax1.xaxis_date()
ax1.set_ylabel("MSFT", color = 'k')
ax2 = ax1.twinx()
ax2.plot(date, price_aapl, color = "royalblue")
ax2.set_ylabel("AAPL", color = "royalblue")
plt.title('Microsoft and Apple, past twenty years')
```

```python
# Histogram of daily prices ("distribution")
df_msft = df[df.ticker == 'MSFT']
df_msft['pct_chg'] = df_msft['price.close'].pct_change(periods = 1)
```

```python
df_msft.head()
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])
ax.set_title('MSFT Daily Price Changes')
ax.hist(df_msft.pct_chg, bins = 50);
```
